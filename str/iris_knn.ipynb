```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

iris = load_iris(as_frame=True)

X = iris.data
y = iris.target
target_names = iris.target_names

display(X.head())
display(y.head())
print("features shape:", X.shape)
print("classes name:", target_names)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print("train data:", X_train.shape, "test data:", X_test.shape)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(f"accuracy: {acc:.3f}")

cm = confusion_matrix(y_test, y_pred, labels=np.unique(y))
print("Confusion Matrix (raw):\n", cm)

index_names = [f"true_{name}" for name in target_names]
column_names = [f"pred_{name}" for name in target_names]

cm_df = pd.DataFrame(cm, index=index_names, columns=column_names)

print("\nConfusion matrix (as DataFrame):")
display(cm_df)
```
